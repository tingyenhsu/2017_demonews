###自由時報

```python
today = time.strftime("%Y_%m_%d_%H_%M")

temp=[]
for n in page_n:
    url = "http://news.ltn.com.tw/list/politics?page=" + str(n)
    data = urllib.urlopen(url).read()
    soup = BeautifulSoup(data)
    mydivs = soup.find_all(id='newslistul')
    print mydivs
    if len(mydivs)!=0:
        temp.append(mydivs)    
    else:
        print "FINISH"
        break

# 依照日期存成txt檔
fout = open('raw/LTN/'+today+'raw.txt', 'w')
for t in temp:
    fout.writelines(str(t))
fout.close()
```

###蘋果日報

```python
today = time.strftime("%Y_%m_%d_%H_%M")

temp=[]
for n in page_n:
    url = "http://www.appledaily.com.tw/realtimenews/section/politics/" + str(n)
    data = urllib.urlopen(url).read()
    soup = BeautifulSoup(data)
    mydivs = soup.find_all("li", class_="rtddt")
    if len(mydivs)!= 0:#第一個：<h2> 優質豐富的新聞媒體</h2>
        temp.append(mydivs)
#         print mydivs[1:]         
    else:
        print "FINISH"
        break

# 依照日期存成txt檔
fout = open('raw/Apple/'+today+'raw.txt', 'w')
for t in temp:
    fout.writelines(str(t))
fout.close()
```
